The official CIC-IDS-2018 `Processed Traffic Data for ML Algorithms` downloaded from AWS, while already labeled, does not contain topology vectors like `Dst IP`, `Src IP` and `Src Port`. Because of this omission, network topology analytics like **ip2vec** would be severely constrained, essentially operating merely on **two token features** (**Protocol** and **Dst Port**).

The Destination Port and Protocol are indeed crucial invariants for detecting attack iteration types. However, raw initial **Source Ports and IP addresses would simply overfit the dataset** in simulated scenarios.

The **Source Port** is typically an **ephemeral** port, selected dynamically by the TCP/IP stack from an arbitrary range.
It is assigned solely for the purpose of **uniquely managing multiple parallel connections**.
Since these ports are randomized dynamically and lack semantic meaning regarding the type of service, they act as poor indicators for realistic traffic analysis and flow attribution.

Training on the IPs of an indiscriminate simulation would always yield nonsensical findings. **However, a real scenario leveraging Synthetic Augmented Data would be fundamentally different:** by re-injecting carefully mapped, verifiable IPs across the datasets, we can successfully capture realistic, foundational **relationships between IPs on the global network via algorithms like ip2vec.**
